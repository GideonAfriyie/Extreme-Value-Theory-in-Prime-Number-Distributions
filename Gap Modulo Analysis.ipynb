```python
with open('first_1000000_primes.csv') as f:
    primes = [int(line.strip()) for i, line in enumerate(f) if i > 0]  # skip header

gaps = [primes[i+1] - primes[i] for i in range(len(primes)-1)]

# Computing modulo 6 of each gap
gap_mod_6 = [gap % 6 for gap in gaps]

# Frequency count
import collections
mod6_counts = collections.Counter(gap_mod_6)

# Bar plot
import matplotlib.pyplot as plt

mods = list(mod6_counts.keys())
counts = [mod6_counts[m] for m in mods]

plt.bar(mods, counts, color='teal')
plt.title("Distribution of Prime Gaps mod 6")
plt.xlabel("Gap % 6")
plt.ylabel("Frequency")
plt.xticks(mods)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
