```python
import csv

# Loading primes from CSV
primes = []
with open('first_1000000_primes.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)  # Skip header
    for row in reader:
        if row:  # avoid empty rows
            primes.append(int(row[0]))


twin_primes = [(primes[i], primes[i+1]) for i in range(len(primes)-1) if primes[i+1] - primes[i] == 2]
print(f"Total Twin Prime Pairs: {len(twin_primes)}")

# Frequency by range (every 100,000)
block_size = 100000
blocks = [0] * (len(primes) // block_size)

for p1, _ in twin_primes:
    idx = p1 // block_size
    if idx < len(blocks):
        blocks[idx] += 1


plt.figure(figsize=(10, 5))
plt.bar([f"{i*100}k" for i in range(len(blocks))], blocks, width=0.6)
plt.xticks(rotation=45)
plt.title("Twin Primes Frequency per 100k Range")
plt.xlabel("Prime Range")
plt.ylabel("Twin Primes Count")
plt.tight_layout()
plt.show()
```
