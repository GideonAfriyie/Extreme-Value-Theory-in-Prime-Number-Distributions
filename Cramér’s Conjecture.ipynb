```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Load primes
with open('first_1000000_primes.csv') as f:
    primes = [int(line.strip()) for i, line in enumerate(f) if i > 0]  # skip header

# Gaps
gaps = [primes[i+1] - primes[i] for i in range(len(primes)-1)]
log_squared = [math.log(p)**2 for p in primes[:-1]]

# Plot
plt.figure(figsize=(12, 6))
plt.plot(primes[:-1], gaps, label='Observed Gaps', alpha=0.6)
plt.plot(primes[:-1], log_squared, label='log(p)^2 (Cramér)', color='red', linestyle='--')
plt.title("Cramér’s Conjecture vs. Observed Prime Gaps")
plt.xlabel("Prime Number")
plt.ylabel("Gap")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

ratios = [gaps[i] / log_squared[i] for i in range(len(gaps))]

plt.hist(ratios, bins=100, color='skyblue')
plt.title('Distribution of Gaps / log(p)^2')
plt.xlabel('Ratio')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
