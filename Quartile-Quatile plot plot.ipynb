```python
import math
import matplotlib.pyplot as plt
import scipy.stats as stats


with open('first_1000000_primes.csv') as f:
    primes = [int(line.strip()) for i, line in enumerate(f) if i > 0]


gaps = [primes[i+1] - primes[i] for i in range(len(primes)-1)]


log_squared = [math.log(p)**2 for p in primes[:-1]]

ratios = [gaps[i] / log_squared[i] for i in range(len(gaps))]

stats.probplot(ratios, dist="norm", plot=plt)
plt.title("Q-Q Plot of Gap/Log^2 Ratios vs. Normal Distribution")
plt.grid(True)
plt.show()

plt.hist(ratios, bins=100, color='skyblue', edgecolor='black')
plt.title("Histogram of Gap / log²(p) Ratios")
plt.xlabel("Ratio")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

plt.boxplot(ratios, vert=False)
plt.title("Boxplot of Gap / log²(p) Ratios")
plt.grid(True)
plt.show()

```
